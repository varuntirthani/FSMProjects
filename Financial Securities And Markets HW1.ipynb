Loan Accounting (60 pts)

We consider a loan with principal P and N amortizing payments. The interest amount is computed from the remaining principal with a simple rate R.

The amount paid to refund the principal at the end of each period is called the annuity amount or simply the annuity.

We note $C_{k}$ the cash amount paid on the kth period and $A_{k}$ the corresponding amortizing amount.

As such we have P = $\sum \limits_{k=1} ^{N}$ $A_{k}$ with $C_{k} = A_{k} + I_{k}$. The remaining capital is $P_{k}$ = P − $\sum \limits_{i=1} ^{k-1}$ $A_{i}$

<br> P = $\sum \limits_{k=1} ^{N}$ $A_{k}$ ... Equation 1

<br> $P_{k}$ = P − $\sum \limits_{i=1} ^{k-1}$ $A_{i}$ .. Equation 3

<br> In Equation 3, if we substitute k as n, we get,
<br> $P_{n}$ = P − $\sum \limits_{i=1} ^{n-1}$ $A_{i}$
<br> But from Equation 1, we know that P = $\sum \limits_{k=1} ^{N}$ $A_{k}$.
<br> Hence if we combine these 2 equations, we get,
<br> $P_{n}$ = $\sum \limits_{k=1} ^{N}$ $A_{k}$ − $\sum \limits_{i=1} ^{n-1}$ $A_{i}$

<br> i.e $P_{n}$ = $A_{n}$ (The 2nd and 3rd terms of the RHS cancel each other out as they are identical)

<br> Now, from Equation 2, we have $C_{k} = A_{k} + I_{k}$
<br> In Equation 2, if we substitute k as n, we get,
<br> $C_{n} = A_{n} + I_{n}$
<br> But, we know that $I_{n}$ = R*$P_{n}$ (Simple Interest On Remaining Capital)
<br> Since, we have already proven above that $P_{n}$ = $A_{n}$, hence $I_{n}$ = R * $A_{n}$
<br> i.e $C_{n}$ = $A_{n}$ + R X $A_{n}$

d) Write a Python notebook that build an annuity table and compute the annuity amount and the prepayment amount at each annuity where the frequency is assumed to be monthly while other parameters are user inputs.

```python
def calculate_annuity(P, r, N):
    """
    Calculate the annuity amount for a loan.
    
    Parameters:
        P (float): Principal amount of the loan.
        r (float): Annual interest rate (decimal).
        N (int): Total number of payments.
        
    Returns:
        float: Annuity amount.
    """
    R = r / 12  # Monthly periodic interest rate
    A = (P * R) / (1 - (1 + R)**-N)
    return A

def calculate_remaining_balance(A, r, n):
    """
    Calculate the remaining balance of a loan after 'n' payments.
    
    Parameters:
        A (float): Annuity amount of the loan.
        r (float): Annual interest rate (decimal).
        n (int): Number of payments made.
        
    Returns:
        float: Remaining balance.
    """
    R = r / 12  # Monthly periodic interest rate
    remaining_balance = A * ((1 - (1 + R)**-n) / R)
    return remaining_balance

def build_annuity_table(P, r, N):
    """
    Build an annuity table.
    
    Parameters:
        P (float): Principal amount of the loan.
        r (float): Annual interest rate (decimal).
        N (int): Total number of payments.
        
    Returns:
        list of tuples: Annuity table containing annuity amount and remaining balance at each payment.
    """
    annuity = calculate_annuity(P, r, N)
    annuity_table = []
    remaining_balance = P
    for n in range(1, N + 1):
        remaining_balance -= annuity - (remaining_balance * (r / 12)) # Removing the interest component from the annuity
        annuity_table.append((annuity, remaining_balance))
    return annuity_table

def main():
    # User inputs
    P = float(input("Enter principal amount of the loan: "))
    r = float(input("Enter annual interest rate (in decimal): "))
    N = int(input("Enter total number of payments: "))
    
    # Build annuity table
    annuity_table = build_annuity_table(P, r, N)
    
    # Print annuity table
    print("\nAnnuity Table:")
    print("Payment Number | Annuity Amount | Remaining Balance")
    print("---------------------------------------------------")
    for i, (annuity, remaining_balance) in enumerate(annuity_table, start=1):
        print(f"Payment {i:5d}      | {annuity:13.2f} | {remaining_balance:17.2f}")
    
    # Calculate prepayment amount
    n_prepayment = int(input("\nEnter the number of payments you want to skip (prepay): "))
    prepayment_amount = calculate_remaining_balance(calculate_annuity(P, r, N), r, n_prepayment)
    print(f"\nPrepayment amount after {N - n_prepayment} payments: {prepayment_amount:.2f}")

if __name__ == "__main__":
    main()
```

Simulation Of Option Hedging (70 pts)
<br> A standard way to simulate paths of stochastic processes over an internal \[0,T\] is to consider a number of time step N, then set δt = T/N and then simulate $W_{t}$ at discrete time $t_{i}$ = iδt.
<br> To simulate a Brownian motion (BM) from $t_{i}$ to $t_{i+1}$ we use that $Wt_{i+1}$ = $Wt_{i}$+($Wt_{i+1}$−$Wt_{i}$), with the property that $Wt_{i+1}$ − $Wt_{i}$ is independent of $Wt_{i}$ together with $Wt_{i+1}$ − $Wt_{i}$ ∼ $\sqrt{t_{i+1} − t_{i}}Z_{i}$ = $\sqrt{δt}Z_{i}$ with $Z_{i}$ ∼ N(0, 1). As such, we can simulate $Z_{1}$, ...,$Z_{N}$ and set $Wt_{i}$ = $Wt_{i−1}$ + $\sqrt{δt}Z_{i}$

Q1) Simulation of Brownian Motion Paths (5 pts)
<br> a) Write a function that simulate paths of Brownian Motion

```python
import numpy as np

def simulate_brownian_motion(T, N, num_paths):
    """
    Simulate paths of Brownian motion.
    
    Parameters:
        T (float): Time horizon.
        N (int): Number of time steps.
        num_paths (int): Number of paths to simulate.
        
    Returns:
        numpy.ndarray: Array of shape (N+1, num_paths) containing simulated paths of Brownian motion.
    """
    # Initialize array to store Brownian motion paths
    brownian_paths = np.zeros((N+1, num_paths))
    
    # Calculate time step
    dt = T / N
    
    # Simulate paths
    for i in range(1, N+1):
        # Generate random normal variables for each time step
        Z = np.random.normal(0, 1, num_paths)
        # Update Brownian motion paths using the formula W_{t_i} = W_{t_{i-1}} + sqrt(dt) * Z_i
        brownian_paths[i] = brownian_paths[i-1] + np.sqrt(dt) * Z
    
    return brownian_paths

# Example usage:
T = 1.0  # Time horizon
N = 100  # Number of time steps
num_paths = 5  # Number of paths to simulate

brownian_paths = simulate_brownian_motion(T, N, num_paths)
print("Simulated paths of Brownian motion:")
print(brownian_paths)
```

Q2) Simulation of Brownian Motion Paths (5 pts)
<br> We assume a Black-Scholes underlying with $dS_{t}$ = $μS_{t}dt$ + $σS_{t}dW_{t}$. As a baseline, we assume $S_{0}$ = 100, μ = 3%, r = 5%, σ = 25%
<br> To simulate the underlying $S_{t}$ we notice that $S_{ti+1}$ = $S_{ti}e^{μ−\frac{1}{2}σ^{2}(t_{i+1}−t_{i})+σ(W_{ti+1}−W_{ti})}$ = $S_{ti}e^{(μ−\frac{1}{2}σ^{2})δt + σ\sqrt{δt}Z_{t+1}}$
<br> Repeat the previous question but on $S_{t}$ this time

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_underlying(S0, T, N, num_paths, mu, sigma):
    """
    Simulate paths of the underlying asset according to the Black-Scholes model.
    
    Parameters:
        S0 (float): Initial price of the underlying asset.
        T (float): Time horizon.
        N (int): Number of time steps.
        num_paths (int): Number of paths to simulate.
        mu (float): Drift coefficient.
        sigma (float): Volatility coefficient.
        
    Returns:
        numpy.ndarray: Array of shape (N+1, num_paths) containing simulated paths of the underlying asset.
    """
    # Initialize array to store underlying asset paths
    underlying_paths = np.zeros((N+1, num_paths))
    
    # Calculate time step
    dt = T / N
    
    # Simulate paths
    for i in range(1, N+1):
        # Generate random normal variables for each time step
        Z = np.random.normal(0, 1, num_paths)
        # Update underlying asset paths using the Black-Scholes formula
        underlying_paths[i] = underlying_paths[i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)
    
    return underlying_paths

# Parameters
S0 = 100  # Initial price of the underlying asset
T = 1.0  # Time horizon
N = 100  # Number of time steps
num_paths = 10000  # Number of paths to simulate
mu = 0.03  # Drift coefficient
sigma = 0.25  # Volatility coefficient

# Simulate underlying asset paths
underlying_paths = simulate_underlying(S0, T, N, num_paths, mu, sigma)

# Calculate mean and standard deviation at each time step
statistics = calculate_statistics(underlying_paths)

# Theoretical values for E(St) and sqrt(Var(St))
t_values = np.linspace(0, T, N+1)
E_St = S0 * np.exp(mu * t_values)
sqrt_Var_St = S0 * np.exp(mu * t_values) * np.sqrt(np.exp(sigma**2 * t_values) - 1)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t_values, statistics[:, 0], label='Sample Mean E(St)', color='blue')
plt.plot(t_values, statistics[:, 1], label='Sample Std Dev sqrt(Var(St))', color='red')
plt.plot(t_values, E_St, label='Theoretical E(St)', linestyle='--', color='green')
plt.plot(t_values, sqrt_Var_St, label='Theoretical sqrt(Var(St))', linestyle='--', color='orange')
plt.title('Simulated Underlying Asset Paths (Black-Scholes Model)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

To simulate the values for the hedging portfolio $P_{t}$ = $Δ_{t}S_{t}$ + $ϕ_{t}B_{t}$, where $Δ_{t}$ represents the number of shares of the underlying asset $S_{t}$ and $ϕ_{t}$ represents the amount of cash invested in the risk-free asset $B_{t}$, we can follow these steps:

```python
import numpy as np
from scipy.stats import norm

def simulate_underlying(S0, T, N, num_paths, r, sigma):
    """
    Simulate paths for the underlying asset S_t according to the Black-Scholes model.
    
    Parameters:
        S0 (float): Initial price of the underlying asset.
        T (float): Time horizon.
        N (int): Number of time steps.
        num_paths (int): Number of paths to simulate.
        r (float): Constant risk-free interest rate.
        sigma (float): Volatility coefficient.
        
    Returns:
        numpy.ndarray: Array of shape (N+1, num_paths) containing simulated paths of the underlying asset.
    """
    dt = T / N
    sqrt_dt = np.sqrt(dt)
    # Initialize array to store paths
    paths = np.zeros((N+1, num_paths))
    # Set initial values
    paths[0] = S0
    # Simulate paths
    for i in range(1, N+1):
        Z = np.random.normal(size=num_paths)
        paths[i] = paths[i-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * sqrt_dt * Z)
    return paths

def calculate_option_prices(underlying_paths, K, T, N, r, sigma):
    """
    Calculate option prices for each time step based on the simulated paths of the underlying asset.
    
    Parameters:
        underlying_paths (numpy.ndarray): Array of shape (N+1, num_paths) containing simulated paths of the underlying asset.
        K (float): Strike price of the option.
        T (float): Time horizon.
        N (int): Number of time steps.
        r (float): Constant risk-free interest rate.
        sigma (float): Volatility coefficient.
        
    Returns:
        numpy.ndarray: Array of shape (N+1, num_paths) containing option prices for each time step.
    """
    dt = T / N
    option_prices = np.zeros_like(underlying_paths)
    for i in range(N+1):
        option_prices[i] = np.maximum(underlying_paths[i] - K, 0)
    return option_prices

def calculate_deltas(underlying_paths, K, T, N, r, sigma):
    """
    Calculate deltas for each time step using the Black-Scholes delta formula.

    Parameters:
        underlying_paths (numpy.ndarray): Array of shape (N+1, num_paths) containing simulated paths of the underlying asset.
        K (float): Strike price of the option.
        T (float): Time horizon.
        N (int): Number of time steps.
        r (float): Constant risk-free interest rate.
        sigma (float): Volatility coefficient.

    Returns:
        numpy.ndarray: Array of shape (N+1, num_paths) containing deltas for each time step.
    """
    dt = T / N
    deltas = np.zeros_like(underlying_paths)
    for i in range(N + 1):
        S = underlying_paths[i]
        denominator = sigma * np.sqrt(T - i * dt)
        if denominator == 0:
            d1 = 0  # Set d1 to 0 if the denominator is 0
        else:
            d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * (T - i * dt)) / denominator
        deltas[i] = norm.cdf(d1)
    return deltas


def simulate_hedging_portfolio(S0, K, T, N, num_paths, r, sigma):
    """
    Simulate values for the hedging portfolio.
    
    Parameters:
        S0 (float): Initial price of the underlying asset.
        K (float): Strike price of the option.
        T (float): Time horizon.
        N (int): Number of time steps.
        num_paths (int): Number of paths to simulate.
        r (float): Constant risk-free interest rate.
        sigma (float): Volatility coefficient.
        
    Returns:
        numpy.ndarray: Array of shape (N+1, num_paths) containing simulated values for the hedging portfolio.
    """
    # Initialize array to store values for the hedging portfolio
    hedging_portfolio_values = np.zeros((N+1, num_paths))
    
    # Calculate time step
    dt = T / N
    
    # Simulate paths for the underlying asset S_t according to the Black-Scholes model
    underlying_paths = simulate_underlying(S0, T, N, num_paths, r, sigma)
    
    # Simulate paths for the risk-free asset B_t
    risk_free_paths = np.exp(r * np.arange(N+1) * dt)
    
    # Calculate option price C_t for each time step t based on the simulated paths of S_t
    option_prices = calculate_option_prices(underlying_paths, K, T, N, r, sigma)
    
    # Calculate delta Δ_t for each time step t using the Black-Scholes delta formula
    deltas = calculate_deltas(underlying_paths, K, T, N, r, sigma)
    
    # Simulate values for the hedging portfolio
    for i in range(N+1):
        # Update the hedging portfolio value P_t = Δ_t S_t + φ_t B_t
        hedging_portfolio_values[i] = deltas[i] * underlying_paths[i] + (option_prices[i] - deltas[i] * underlying_paths[i])
    
    return hedging_portfolio_values

# Parameters
S0 = 100  # Initial price of the underlying asset
K = S0  # Strike price of the option
T = 1.0  # Time horizon
N = 365  # Number of time steps (daily simulation)
num_paths = 1000  # Number of paths to simulate
r = 0.05  # Constant risk-free interest rate
sigma = 0.25  # Volatility coefficient

# Simulate values for the hedging portfolio
hedging_portfolio_values = simulate_hedging_portfolio(S0, K, T, N, num_paths, r, sigma)
```

<br> 3) Determine the volatility value where the option price matches the delta-adjusted portfolio value, indicating zero profit or loss.
<br> 4) Repeat this process for multiple paths of $S_{t}$ to get a distribution of B/E volatility values.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Function to calculate option price and delta
def calculate_option_price_and_delta(S0, K, T, r, sigma):
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    option_price = S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    delta = norm.cdf(d1)
    return option_price, delta

# Given parameters
S0 = 100  # Initial stock price
K = 100  # Strike price (ATM)
T = 1  # Time to expiry
r = 0.05  # Risk-free rate

# Number of paths
num_paths = 10000

# Range of volatility values to test
sigma_values = np.linspace(0.01, 1, 100)

# List to store B/E volatility values
be_volatilities = []

# Iterate over paths
for _ in range(num_paths):
    # Generate a random path for St (you can replace this with your actual data)
    St_path = np.random.normal(S0, 0.2, 252)  # Example path with mean S0 and standard deviation 0.2
    
    # Calculate option price and delta for each volatility value
    for sigma in sigma_values:
        option_price, delta = calculate_option_price_and_delta(S0, K, T, r, sigma)
        # Check if option price matches delta-adjusted portfolio value
        if np.isclose(option_price, delta * St_path[-1]):
            be_volatilities.append(sigma)
            break

# Plot histogram of B/E volatility values
plt.hist(be_volatilities, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title('Distribution of Break-Even (B/E) Volatility')
plt.xlabel('B/E Volatility')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
